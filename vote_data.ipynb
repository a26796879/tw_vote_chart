```python
df = pd.read_excel(r'C:\Users\user\Downloads\直轄市議員\縣表3-2-300(桃園市).xls')
df.iloc[4,0] = df.iloc[4,0].split('\u3000')[1]
candidate_col = len(df.columns)-8
count_row = df.shape[0]
change_name = {'107年桃園市議會議員選舉第1選舉區候選人在各投開票所得票數一覽表':'village_name','Unnamed: 1':'村里別','Unnamed: 2':'投開票所別'}
for i in range(3,33-8):
    voter = df.iloc[1, i].split('\n')
    #print(voter)
    df.iloc[0, i] = voter[0]
    df.iloc[1, i] = voter[1]
    df.iloc[2, i] = voter[2]
    change_name['Unnamed: '+str(i)] = i-2
for i in range(candidate_col,33):
    voter = df.iloc[0, i].split('\n')
    #print(voter)
    df.iloc[0, i] = voter[1]
    df.iloc[1, i] = voter[3]
    change_name['Unnamed: '+str(i)] = df.iloc[0, i]
df.iloc[1, 0] = "候選人"
df.iloc[2, 0] = "所屬黨派"
df = df.rename(change_name, axis=1) 

for i in range(5,count_row):
    df.iloc[i, 0] = df.iloc[4, 0] + df.iloc[i, 1]
df2 = df.groupby(['village_name'], as_index=False).sum()
df2 = df2.drop(columns=['村里別','投開票所別'])
df2.to_excel('learn_df.xls', index = False)
```
